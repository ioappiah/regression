<code>Next, we'll want to consider the effect of studytime in hours to the possibility of student performance (i.e.,Pass or Fail).</code>

```python
df['studytime'].describe()
```

<code>Observation :</code> 
</p>75% of students have study times below 2 hours, with maximum study time hours being 4 hours.

    Can we claim that students require a median study time of **less than 3 hours** to achieve exam success?. Let's subject this hypothesis to the test. 
    
Let's assumption that students need at least 3 hours of studytime to achieve exams success.
- **Null Hypothesis (H₀):** Median study time ≥ 3 hours  
- **Alternative Hypothesis (H₁):** Median study time < 3 hours  

```python
import scipy.stats as sts
```

```python
# Hypothesized median (H₀: median >= 3)
median = 3  

sample_data = np.array(df['studytime'])  

test_stat, sample_p_value = sts.wilcoxon(
    x = sample_data - median, 
    alternative = 'less'       # One-tailed: H₁: median < 3
)

print('Test statistic:', test_stat)
print('Research probability value (p-value):', sample_p_value)

# Decision rule
alpha = 0.05
if sample_p_value < alpha:
    print("Reject H₀: Median study time is significantly less than 3 hours (p < 0.05).")
else:
    print("Fail to reject H₀: No evidence that median study time is < 3 hours.")
```

#### <span style="font-family: Candara; text-align: left;font-size:17.25px; letter-spacing: 0.85px;"><b>2.2 Data Preprocessing</b></span>
  * <p>2.2.1 Checking for Missing Values</p>

```python
# Find the number of null values in each columns
df.isnull().sum()
```

**Observation:**

No missing values were identified, indicating that the dataset is clean.

**2.2.2 Encoding**

```python
# Let's take a look at the data dtypes in the data set
df.dtypes
```

**Next, we will list all non-numeric columns (categorical columns) in the dataset.**

```python
categorical_columns = [df.columns[index] for index, dtype in enumerate(df.dtypes) if dtype == 'object']

categorical_columns
```

**Next, let's examine the data types in each non-numeric column by identifying the unique values within each variable.**

```python
for column in categorical_columns:
    print(f'{column}: {df[column].unique()}')
```

**Observation:**

* Many features are binary variables with only two distinct values; for these, we will apply label encoding.
* Features with more than two categories, such as `Mjob`, `Fjob`, `reason`, and `guardian`, will be transformed using one-hot encoding to convert them into numeric format.

**2.2.2.1: One-Hot Encoding**

For variables that have an "other" category value, we will apply one-hot encoding and append unique labels to these values to avoid duplication and ensure distinct representation.

```python
df['Mjob'] = df['Mjob'].apply(lambda x: 'm_' + x)
df['Fjob'] = df['Fjob'].apply(lambda x: 'f_' + x)
df['reason'] = df['reason'].apply(lambda x: 'r_' + x)
df['guardian'] = df['guardian'].apply(lambda x: 'g_'+ x)
```

```python
# Take a look at the data again to check if the function worked properly
df.head()
```

Next, let's create a DataFrame of dummy variables using one-hot encoding for the categorical features with more than two values.

```python
# Create a new dataframe called dummies with the concatenate dummies
dummies = pd.concat([pd.get_dummies(df['Fjob']), pd.get_dummies(df['Mjob']), pd.get_dummies(df['reason']), pd.get_dummies(df['guardian'])], axis = 1)
dummies.head()
```

Next, let’s concatenate the dummies DataFrame to the original DataFrame after dropping the variables that were one-hot encoded.

```python
df = pd.concat([df, dummies], axis = 1)

df.drop(['Mjob', 'Fjob', 'reason', 'guardian'], axis = 1, inplace = True)
```

```python
# Take a look at the data
df.head(3)
```

We will apply Label Encoding to the binary categorical features (i.e., variables with two distinct values) in the dataset.

**2.2.2.2 Label Encode**

```python
encoder = LabelEncoder()

for column in categorical_columns:
    if column not in ['Mjob', 'Fjob', 'reason', 'guardian']:
        df[column] = encoder.fit_transform(df[column])
```

```python
df.head(4)
```

**2.3 Scaling**

```python
X = df.drop(['G3'], axis = 1)
y = df['G3']
```

```python
X.head(4)
```

We'll then scale X with **sklearn- StandardScaler**, which will give every feature unit a variance

```python
# scaling X
scaler = StandardScaler()
X = pd.DataFrame(scaler.fit_transform(X), columns = X.columns)
```

```python
X.head(4)
```

```python
y.head()
```

#### <span style="font-family: Candara; text-align: left;font-size:17.25px; letter-spacing: 0.5px;"><b>Model Building</b>
**Training a model**
> We will split our **feature set (X)** into a **70% training set** and a **20% testing set**. The training set will be used to build the regression model, while the testing set will be used to evaluate its performance.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20)
```

**1. <p>Linear Regression</p>**

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

**Evaluation (Predicting on the Test Data)**

```python
print(f'Model R2: {model.score(X_test, y_test)}')
```

**Insight**:

We achieved an **R² score of 0.77** with the linear regression model, which indicates a strong model performance. This means that, based on all the available features, the model can explain approximately **77% of the variation** in the final course grades, providing a fairly accurate estimate of students’ final scores.

**2. <p>XGBoost Regressor (Gradient Boosting Method)</p>**

```python
from xgboost import XGBRegressor

xgb = XGBRegressor(n_estimators = 500, learning_rate = 0.05, random_state = 42)
xgb.fit(X_train, y_train)
```

**Evaluation (Predicting on the Test Data)**

```python
from sklearn.metrics import mean_squared_error, r2_score

y_pred = xgb.predict(X_test)

print("XGBoost RMSE:", mean_squared_error(y_test, y_pred))
print("R² Score:", r2_score(y_test, y_pred))
```

`Models R² Explanation`:
1. **R² = 0.852** means of our model explains about 89.6% of the variation in students’ final grades (G3) based on the features we used (e.g., study time, absences, past grades, parental education, etc.). In other words, almost 9 out of 10 differences in students’ grades can be predicted by our model from the input data.
2. Again, using the features in the dataset, our model can predict a student’s final grade with about **90% accuracy** in terms of explaining why grades differ. This means that, if we look at two students who got different grades, the model can explain most of the reasons for that difference using the given data — such as study habits, prior grades, and attendance.
